### 1. 可微性


#### 1.1. 重要定义

- **二元函数的偏导**：设函数$z=f(x,y)$，当极限$$\lim_{\Delta x \to 0} \frac{f(x+\Delta x,y)-f(x,y)}{\Delta x}$$存在时，称此极限为$f(x,y)$在点$(x,y)$关于$x$的偏导数，记作$\frac{\partial f}{\partial x}$：$$\frac{\partial f}{\partial x}=\lim_{\Delta x \to 0} \frac{f(x+\Delta x,y)-f(x,y)}{\Delta x}$$同理$$\frac{\partial f}{\partial y}=\lim_{\Delta y \to 0} \frac{f(x,y+\Delta y)-f(x,y)}{\Delta y}$$


- **全微分**：设函数$z=f(x,y)$在点$(x,y)$的某邻域内有定义，对于点$P(x,y)=(x_0+Δx,y_0+Δy)$，若函数$f(x,y)$在点$(x_0,y_0)$的增量$$Δz=f(x_0+Δx,y_0+Δy)-f(x_0,y_0)$$可表示为：$$Δz=AΔx+BΔy+o(\sqrt{Δx^2+Δy^2})$$其中$A,B$为常数，$o(\sqrt{Δx^2+Δy^2})$是当$(Δx,Δy)→(0,0)$时，$Δz$相对于$\sqrt{Δx^2+Δy^2}$的高阶无穷小，则称函数$f(x,y)$在点$(x_0,y_0)$可微，且$AΔx+BΔy$称为函数$f(x,y)$在点$(x_0,y_0)$的全微分，记作$dz$：$$dz=\frac{\partial f}{\partial x}dx+\frac{\partial f}{\partial y}dy$$全微分存在的充要条件是：$$\frac{Δz-AΔx-BΔy}{\sqrt{Δx^2+Δy^2}}→0$$

- **多元函数中连续、偏导存在、可微、偏导连续的关系**：如下图所示也可以简述为：

    - **偏导数存在且连续**，则函数可微
    - 函数可微，则 ①**偏导数存在**；②**函数连续**
    - **函数可微**，则**任意方向导数**存在



- **切平面**：设函数$z=f(x,y)$在点$(x_0,y_0)$可微，则函数在该点的切平面方程为：$$z-f(x_0,y_0)=\frac{\partial f}{\partial x}(x-x_0)+\frac{\partial f}{\partial y}(y-y_0)$$

#### 1.2. 重要定理

- **可微的充要条件**：函数$f(x,y)$在点$(x_0,y_0)$可微的充要条件是：$$\frac{Δz-AΔx-BΔy}{\sqrt{Δx^2+Δy^2}}→0$$

- **偏导数存在且连续**，则函数可微

- 函数可微，则 ①**偏导数存在**；②**函数连续**


- **函数可微**，则**任意方向导数**存在



### 2. 复合函数微分法


#### 2.1. 重要定义

- **二元函数的链式法则**：设函数$z=f(u,v)$在点$(u,v)$**可微**，函数$u=u(x,y),v=v(x,y)$在点$(x,y)$可微，则复合函数$z=f(u(x,y),v(x,y))$在点$(x,y)$可微，且有：$$\frac{\partial z}{\partial x}=\frac{\partial z}{\partial u}\frac{\partial u}{\partial x}+\frac{\partial z}{\partial v}\frac{\partial v}{\partial x}$$ $$\frac{\partial z}{\partial y}=\frac{\partial z}{\partial u}\frac{\partial u}{\partial y}+\frac{\partial z}{\partial v}\frac{\partial v}{\partial y}$$注意，上述条件可以是$u(x,y),v(x,y)$存在偏导数，但是$z=f(x,y)$的 **可微条件** 不可缺少。


### 3. 方向导数与梯度

#### 3.1. 重要定义


 - **二元函数的方向导数**：设二元函数$f(x,y)$ 在点 $P_0(x_0,y_0)$ 的某邻域 $U(P_0)\subset\mathbb{R}^2$内有定义, $\vec{l}$ 为从$P_0$出发的射线, 点$P(x,y)$为$\vec{l}$上且含于$U(P_0)$内的任一点，以$\rho$表示$P$与$P_0$两点间的距离. 若极限
$$\begin{align*}\lim_{\rho\to0^+}\frac{f(P)-f(P_0)}{\rho}&=\lim_{\rho\to0^+}\frac{f(x,y)-f(x_0,y_0)}{\rho}\\&=\lim_{\rho\to0^+}\frac{f(x_0+\rho\cos\alpha,y_0+\rho\sin\alpha)-f(x_0,y_0)}{\rho}\end{align*}$$
$\ \ \quad$ 存在时，称这个极限为函数 $f$在点$P_0(x_0,y_0)$沿方向$\vec{l}$的方向导数,记作$f_{\vec{l}}(x_0,y_0)$ 或 $\frac{\partial f}{\partial \vec{l}}\bigg|_{(x_0,y_0)}$


- **梯度和模**：设函数$z=f(x,y)$在点$(x_0,y_0)$可微，函数$f(x,y)$在点$(x_0,y_0)$的梯度为：$$\nabla f(x_0,y_0)=(\frac{\partial f}{\partial x},\frac{\partial f}{\partial y})|_{(x_0,y_0)}$$函数$f(x,y)$在点$(x_0,y_0)$的梯度模为：$$|\nabla f(x_0,y_0)|=\sqrt{(\frac{\partial f}{\partial x})^2+(\frac{\partial f}{\partial y})^2}|_{(x_0,y_0)}$$

- **可微函数的方向导数**：设函数$z=f(x,y)$在点$(x_0,y_0)$可微，$\vec{l}$为任意方向向量，则函数$f(x,y)$在点$(x_0,y_0)$沿方向$\vec{l}$的方向导数为：$$\frac{\partial f}{\partial \vec{l}}\bigg|_{(x_0,y_0)}=\text{gard}f(x_0,y_0)\cdot\vec{l}=(\frac{\partial f}{\partial x}\cos\alpha+\frac{\partial f}{\partial y}\cos\beta)|_{(x_0,y_0)}$$其中$\cos\alpha,\cos\beta$为$\vec{l}$的方向余弦.
    - 当两向量夹角为$0$时，即$\vec{l}$与$\text{gard}f(x_0,y_0)$同向，方向导数最大，为$|\text{gard}f(x_0,y_0)|$
    - 当两向量夹角为$\pi$时，即$\vec{l}$与$\text{gard}f(x_0,y_0)$反向，方向导数最小，为$-|\text{gard}f(x_0,y_0)|$

### 4. 泰勒公式与极值问题



#### 4.1. 重要定义


- **连续的高阶偏导数次序无关**：设函数$z=f(x,y)$在点$(x_0,y_0)$的某邻域内有定义，若函数$f(x,y)$的偏导数$\frac{\partial^2 f}{\partial x\partial y},\frac{\partial^2 f}{\partial y\partial x}$在该邻域内**连续**，则这些偏导数的值与次序无关，即：$$\frac{\partial^2 f}{\partial x\partial y}=\frac{\partial^2 f}{\partial y\partial x}$$推广到高阶也成立。

- **凸区域**：若区域$D$上任意两点的连线都在$D$内，则称$D$为凸区域，也就是对于$D$内任意两点$(x_1,y_1),(x_2,y_2)$，都有：$$P(x_1+\lambda (x_2-x_1),y_1+\lambda (y_2-y_1))\in D,\forall \lambda\in[0,1]$$

- **极值**：设函数$z=f(x,y)$在点$(x_0,y_0)$的某邻域内有定义，若存在$(x_0,y_0)$的某邻域$U(P_0)$，使得$U(P_0)$内的任意点$(x,y)$都有$f(x,y)\leq f(x_0,y_0)$或$f(x,y)\geq f(x_0,y_0)$，则称$f(x_0,y_0)$为函数$f(x,y)$的极大值或极小值，$f(x_0,y_0)$称为函数$f(x,y)$的极值。

- **稳定点**：设二元函数$z=f(x,y)$在点$(x_0,y_0)$有偏导数，且有$f_x|_{(x_0,y_0)}=f_y|_{(x_0,y_0)}=0$，则称$(x_0,y_0)$为函数$f(x,y)$的稳定点

- **Hesse矩阵**：设函数$z=f(x,y)$在点$(x_0,y_0)$的某邻域内有定义，若函数$f(x,y)$在该邻域内有直到$n$阶的连续偏导数，则称函数$f(x,y)$在点$(x_0,y_0)$的Hesse矩阵为：$$H(f(x_0,y_0))=\begin{pmatrix}\frac{\partial^2 f}{\partial x^2}&\frac{\partial^2 f}{\partial x\partial y}\\\frac{\partial^2 f}{\partial y\partial x}&\frac{\partial^2 f}{\partial y^2}\end{pmatrix}|_{(x_0,y_0)}$$

- **二阶连续偏导的极值**：设函数$z=f(x,y)$在点$(x_0,y_0)$的某邻域内有定义，且有**二阶的连续偏导数**，且$(x_0,y_0)$为$f(x,y)$的稳定点，则有：
    - 若$H(f(x_0,y_0))$正定，则$f(x_0,y_0)$为极小值
    - 若$H(f(x_0,y_0))$负定，则$f(x_0,y_0)$为极大值
    - 若$H(f(x_0,y_0))$不定，则$f(x_0,y_0)$不是极值

- **最值点**：设函数$z=f(x,y)$在区域$D$上有定义，若函数$f(x,y)$在$D$上有最大值或最小值，则称$f(x,y)$在$D$上有最值，$f(x,y)$的最值点称为$f(x,y)$的最值点，最值点可能出现在：
    - 稳定点
    - 边界点
    - 无偏导数的点

#### 4.2. 重要定理

- **中值定理**：设函数$z=f(x,y)$在凸开区域$D\subset\mathbb{R}^2$上**连续**，在 $D$ 的所有内点都**可微**，则对于 $D$ 内任意两点 $(x_0,y_0),(x_0+h,y_0+k)$，存在某个$\theta\in(0,1)$，使得：$$f(x_0+h,y_0+k)-f(x_0,y_0)=\frac{\partial f}{\partial x}(x_0+\theta h,y_0+\theta k)h+\frac{\partial f}{\partial y}(x_0+\theta h,y_0+\theta k)k$$
    - **推论**：设函数$f(x,y)$在$D$上存在偏导数，且$f_x=f_y=0$，则$f(x,y)$在$D$上横为常数

- **泰勒定理**：设二元函数$z=f(x,y)$在点$(x_0,y_0)$的某邻域内有定义，若函数$f(x,y)$在该邻域内有直到$n$阶的连续偏导数，则对于该邻域内任意一点$(x+h,y+k)$，存在某个$\theta\in(0,1)$，则有**泰勒公式**：

$$
\begin{align*}
f(x_0+h,y_0+k)=&f(x_0,y_0)+(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})f{(x_0,y_0)}+\frac{1}{2!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^2f{(x_0,y_0)}\\
&+\cdots+\frac{1}{n!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^nf{(x_0,y_0)}+R_n
\end{align*}
$$
$\quad\ \ \ \ $其中$R_n$为余项，有$$R_n=\frac{1}{(n+1)!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^{n+1}f{(x_0+\theta h,y_0+\theta k)}\sim o((\sqrt{h^2+k^2})^n)$$
$\quad\ \ \ \ $其中
$$(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^nf{(x_0,y_0)}=\sum_{i=0}^nC_n^i(h\frac{\partial}{\partial x})^i(k\frac{\partial}{\partial y})^{n-i}f{(x_0,y_0)}$$
