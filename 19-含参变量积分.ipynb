### 含参变量积分

#### 1. 重要定义

- 设函数 $f(x,y)$ 定义在无界区域 $\{(x,y)|x\in I, c≤y<+\infty\}$ 上，且满足$$\Phi(x) = \int_c^{+\infty}f(x, y)dy,\quad x\in I$$则称 $\Phi(x)$ 为定义在 $I$ 上的含参量 $x$ 的无穷限反常积分，简称 **含参量反常积分**。

- **含参量反常积分的一致收敛**：若对于给定的正数 $\epsilon$，存在正数 $N>c$，使得在 $M>N$ 时对于一切 $x\in I$，有$$\left|\int_c^Mf(x, y)dy - \Phi(x)\right|<\epsilon$$或$$\left|\int_M^{+\infty}f(x, y)dy\right|<\epsilon$$成立，
则称含参量反常积分在 $I$ 上 **一致收敛**。

- **一致收敛的柯西准则**：对于一切正数 $\epsilon$，若存在正数 $N>c$，使得在 $A_2>A_1>N$ 时对于一切 $x\in I$，有$$\left|\int_{A_1}^{A_2}f(x, y)dy\right|<\epsilon$$成立，则含参量反常积分在 $I$ 上 **一致收敛**。

- 含参量反常积分 $\int_{c}^{+\infty}f(x,y)dy$ 在 $I$ 上 **一致收敛** 的 **充分必要条件是**：$$\lim_{A\to+\infty} F(A) = 0$$ 其中 $F(A) = \sup_{x\in I}|\int_{A}^{\infty}f(x,y)dy|$，$I$ 为 $x$ 的一个区间。

- **内闭一致收敛**：若对于任意的 $[a,b]\subset I$，含参量反常积分在 $[a,b]$ 上一致收敛，则称其在 $I$ 上内闭一致收敛。

#### 2. 重要定理

- **魏尔斯特拉斯 M 判别法**：设 $g(y)$ 满足 $|f(x,y)|\leq g(y)$，且 $$\int_c^{+\infty}g(y)dy$$收敛，则 $$\int_c^{+\infty}f(x,y)dy$$在 $I$ 上一致收敛。

- **狄利克雷判别法**：若

    - $\int_c^{+\infty}f(x,y)dy$ 对 $x$ 在 $I$ 上一致有界
    
    - 且 $g(x,y)$ 在 $[c,+\infty)$ 上单调趋于 0
    
    则 $$\int_c^{+\infty}f(x,y)g(x,y)dy$$在 $I$ 上一致收敛。

- **阿贝尔判别法**：若

    -  $\int_c^{+\infty}f(x,y)dy$ 在 $I$ 上一致收敛
    
    - 且 $g(x,y)$ 在 $I$ 上一致有界
    
    则 $$\int_c^{+\infty}f(x,y)g(x,y)dy$$在 $I$ 上一致收敛。
