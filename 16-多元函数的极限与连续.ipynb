### 1. 点集


#### 1.1. 重要定义


- **内点**：若存在点 $A$ 的某邻域 $U(A)$，使得 $U(A) \subset E$，则称点 $A$ 是集合 $E$ 的内点。

- **内部**：集合 $E$ 的全体内点称为 $E$ 的内部，记为 $\text{int}(E)$。

- **外点**：若存在点 $A$ 的某邻域 $U(A)$，使得 $U(A) \cap E = \varnothing$，则称点 $A$ 是集合 $E$ 的外点。

- **边界点**：若 $\forall δ>0$，点 $A$ 的任一邻域 $U(A;δ)$ 内既含有集合 $E$ 中的点，又含有不属于集合 $E$ 的点，则称点 $A$ 是集合 $E$ 的边界点，记作 $\partial E$。


- **余集**：集合 $E$ 的全体外点称为 $E$ 的余集，记为 $E^c$。

- **聚点**：若点 $A$ 的任一 **去心** 邻域 $U°(A)$ 内至少含有集合 $E$ 中的一点，则称点 $A$ 是集合 $E$ 的聚点。

    - **注意**：一般开集的内点、边界点都是聚点

- **导集**：集合 $E$ 的全体聚点称为 $E$ 的导集，记为 $E^d$。

- **孤立点**：若点 $A$ 存在某邻域 $U(A)$ 内只含有点 $A$ 本身，则称点 $A$ 是集合 $E$ 的孤立点。

- **开集**：若集合 $E$ 的每一点都是 $E$ 的内点，则称 $E$ 为开集。

    - **注意**：开集的内点、边界点都是聚点

- **闭集**：若集合 $E$ 的所有聚点都属于 $E$，则称 $E$ 为闭集。

- **连通性**:非空开集 $E$ 中的任意两点之间都可以用一条完全含于 $E$ 的有限折线相连接

- **区域**：开区域、闭区域、或者开域及其部分边界点，其中
    - **开区域**：有**连通性**的非空开集称为开区域。
    - **闭区域**：开区域及其边界组成的集合。






- **有界点集**：若存在一个有限的实数 $M$，使得 $E$ 中的每一点都在以原点为中心、以 $M$ 为半径的某个球内，则称 $E$ 为有界点集。

- **直径**：若 $E$ 为有界点集，则 $E$ 中任意两点间的最大距离称为 $E$ 的直径。

- **收敛**：设 $\{P_n\}$ 是点集 $E$ 中的一个点列，若对于给定的任意正数$\epsilon$，总存在正整数 $N$，使得当 $n>N$ 时，有 $P_n\in U(P_0;\epsilon)$，则称点列 $\{P_n\}$ 收敛于点 $P_0$，记作 $\lim_{n \to \infty} P_n = P_0$。

    - **等价性**：$\lim_{n \to \infty} P_n = P_0$ 等价于 $\lim_{n \to \infty} x_n = x_0$，$\lim_{n \to \infty} y_n = y_0$。

#### 1.2. 常见推论

- **孤立点一定是边界点**

- 内点和非孤立的边界点一定是聚点

- 既不是聚点，也不是孤立点的点，则必为外点

- 空集和 $\mathbb{R}^n$ 是既是开集又是闭集

- 区分聚点和内点，其中聚点不一定是内点，内点一定是聚点

    - **线上有聚点，但没有内点**（★★★★★）

- **导集** $E^c$ 是所有**聚点**的集合（包括*内点*和*边界点*）

#### 1.3. 重要定理

- **开集和闭集的关系**：集合$E$为闭集的充分必要条件是: ① $E^c$ 为开集 ② $\partial E \subset E$

- **柯西准则**：点列 $\{P_n\}$ 收敛的充分必要条件是：对于任意给定的 $\epsilon>0$，存在正整数 $N$，使得当 $n,p>N$ 时，有 $d(P_n,P_{n+p})<\epsilon$。

- **闭域套定理**：设 $\{G_n\}$ 是$R^2$一列闭区域，且满足 

    (i). $G_1\supset G_2\supset \cdots \supset G_n \supset \cdots$，

    (ii). $\lim_{n \to \infty} d(G_n) = 0$，

    则存在唯一一点 $P_0$，使得 $P_0\in G_n$，$\forall n = 1,2,\cdots$

- **聚点定理**：设 $E$ 为 $R^2$ 中的无限点集，则 $E$ 中至少有一个聚点。

    - **推论**： 有界无限点列$\{P_n\}$必有收敛子列$\{P_{n_k}\}$

- **有界覆盖定理**：设 $D\supset R^2$ 为一有界闭区域，$\{G_\alpha\}$ 为 $D$ 中的一列开区域，它覆盖了$D$，也就是$D\subset \bigcup_{\alpha} G_\alpha$，则必存在有限个开区域 $G_1,G_2,\cdots,G_m$，使得$D\subset \bigcup_{n=1}^{m} G_n$。

### 2. 二元函数

#### 2.1. 重要定义

- **二元函数**：设$D\subset\mathbb{R}^2$，若按照某种对应法则$f$，$D$中的每一点$P(x,y)$都有唯一确定的实数$z$与之相对应，则称$f$为定义在$D$上的二元函数，记作$$f:D\to\mathbb{R},\quad P\mapsto z$$且称$D$为$f$的定义域。$P\in D$所对应的$z$为$f$在点$P$的函数值，记作$z=f(P)$ 或者$z=f(x,y)$. 全体函数值的集合称为值域，记作$f(D)\subset\mathbb{R}$. 

- **n元函数**：设 $E$ 为$R^n$中的点集，若有某个对应法则 $f$ ,使$E$中每一个点$P(x_1,x_2,...,x_n)$,都有唯一的一个实数$y$与之对应，则称$f$为定义在$E$上的$n$元函数，记作 $$f{: }E\to R, ( x_1, x_2, \cdots , x_n) \mapsto y.$$也常把$n$元函数简写成$$y=f\left(x_1,x_2,\cdots,x_n\right),\left(x_1,x_2,\cdots,x_n\right)\in E$$
或着简称$$y=f\left(P\right),P\in E.$$

- **极限**：设 $f$ 是定义在 $D\subset\mathbb{R}^2$ 上的二元函数，$P_0$ 为 $D$ 上的一个聚点，$A$ 是一个确定的实数。若对任给正数 $\epsilon$，总存在某正数 $\delta$，使得当 $P\in U^\circ(P_0;\delta)\cap D$ 时，都有$$|f(P)-A|<\epsilon$$则称 $f$ 在 $D$ 上当 $P\to P_0$ 时以 $A$ 为极限，记做
$$\lim_{P\to P_0, P\in D}f(P)=A.$$

- **无穷大**：$f:D\to R,P_0$ 为 $D$ 的聚点。且 $\forall M>0$ $\exists\delta>0,\forall P\in D:0<\mid PP_0\mid<\delta$,都有$f(P)>M,$则称$\lim_{P\to P_0}f(P)=+\infty.$


- **重极限**：$$\lim_{(x,y)\to(x_0,y_0)}f(x,y)=A$$

- **累次极限**：$$\lim_{x\to x_0}\lim_{y\to y_0}f(x,y)=A,\quad \lim_{y\to y_0}\lim_{x\to x_0}f(x,y)=A$$

 - **连续**：设f为定义在$D⊂ℝ^2$上的二元函数，$P_0∈D$（它是D的**聚点**或者**孤立点**）。对任意正数$ε$，存在正数$δ$，满足当$P∈U(P_0,δ)∩D$ 时就有 $|f(P)−f(P_0)|<ε$，则称$f$关于$D$在点$P_0$连续。
    - 若为孤立点，则为连续点
    - 若为聚点，则仅当$\lim_{P\to P_0}f(P)=f(P_0)$时为连续点



- **连续函数**：若 $f$ 在 $D$ 上任何点都关于集合 $D$ 连续，则称 $f$ 为 $D$ 上的连续函数.


#### 2.2. 重要推论


- $f:D\to R,P_0$为$D$的聚点.则$$\lim_{P\to P_0,P\in D}f(x,y)=A\Leftrightarrow\lim_{P\to P_0,P\in E}f(x,y)=A$$对于任意$E\subset D$且$P_0$为$E$的聚点成立

- **重极限** 和 **累次极限**没有关系 

#### 2.3. 重要定理




- **二元函数的连续性定理**：设函数 $u=\varphi(x,y), v=\psi(x,y)$ 在点 $(x_0,y_0)$ 的某邻域内有定义且连续，函数 $z=f(u,v)$ 在点 $(u_0,v_0)$ 的某邻域内有定义且连续，且 $u_0=\varphi(x_0,y_0),v_0=\psi(x_0,y_0)$，则复合函数 $z=f(\varphi(x,y),\psi(x,y))$ 在点 $(x_0,y_0)$ 处连续。

- **二元函数的有界性定理**：设函数 $f(x,y)$ 在有界闭区域 $D$ 上连续，则 $f(x,y)$ 在 $D$ 上有界，能取得最大值和最小值。

- **二元函数的闭区间一致连续性定理**：设函数 $f(x,y)$ 在有界闭区域 $D$ 上连续，则 $f(x,y)$ 在 $D$ 上一致连续，即$\forall \epsilon>0$时，$P_1,P_2\in D,\exists \delta>0,|P_1P_2|<\delta,|f(P_1)-f(P_2)|<\epsilon$。

- **二元函数的闭区间介值性定理**：设函数 $f(x,y)$ 在有界闭区域 $D$ 上连续，若$P_1,P_2\in D$，且$f(P_1)\neq f(P_2)$，则对于$f(P_1)$与$f(P_2)$之间的任意数 $A$，都存在 $P_3\in D$，使得$f(P_3)=A$。

- **二元函数唯一性定理**：若极限 $\lim_{(x, y) \to (a, b)} f(x, y)$ 存在，则它只有一个极限。
    
- **二元函数局部有界性定理**：若 $\lim_{(x, y) \to (a, b)} f(x, y) = A$，则存在点 $P_{0}(a, b)$ 的某空心邻域 $U^{\circ}(P_{0}; \delta)$，使 $f(x, y)$ 在 $U^{\circ}(P_0;\delta) \cap D$ 上有界。
    
- **二元函数局部保号性定理**：若 $\lim_{(x, y) \to (a, b)} f(x, y) = A > 0$（或 $< 0$），则对任意正数 $r(0 < r < |A|)$，存在 $P_0(a, b)$ 的某空心邻域 $U^\circ(P_0;\delta)$，使得对一切点 $P(x, y) \in U^\circ(P_0;\delta) \cap D$，恒有 $f(x, y) > r > 0$（或 $f(x, y) < -r < 0$）。

### 第十六章做题技巧

- 求二元函数极限：

    - ① 极坐标

    - ② **定义法**（$\epsilon-\delta$）；其中$\delta$用欧氏距离衡量

    - ③ 如果可以取值，要说明：$f$是初等函数，点$(x_0,y_0)$是$f$的定义域的**内点**，从而 **连续**；

-  $\epsilon-\delta$：一般用:
    - **极限存在条件** $|f-A|<\epsilon$ 
    - **柯西收敛准则** $|f_1-f_2|<\epsilon$，
    - **一致连续条件**（普希茨条件）：$|f_1-f_1|≤L|x_1-x_2|$

- 证明题技巧：
    - **三角不等式**：学会一招三角不等式就行了（常用 **极限、柯西收敛、一致连续性**）,例如：$$|f(x,y)-A|≤|f(x,y)-f(x_0,y_0)|+|f(x_0,y_0)-A|$$
    - $\epsilon-\delta$ **技巧**：对于任意的$\epsilon$，都能找到一个$\delta$，你可以随意操控取“任意”的值，构造出需要证明的表达式。

- 讨论连续性，要讨论**间断点**的类型
